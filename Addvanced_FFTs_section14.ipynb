```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# 2D FFT

```python
a = np.mgrid[:3, :3][0]
print(a)
```

```python
print(np.fft.fft2(a))
```

#N Diamentional FFT

```python
a = np.mgrid[:3, :3, :3][0]
print(a)
```

```python
print(np.fft.fftn(a,  axes=(1,2)))
```

```python
print(np.fft.fftn(a,  axes=(0, 1)))
```

```python
[x, y] = np.meshgrid(2* np.pi * np.arange(200) /12,
                    2 * np.pi * np.arange(200) /34)

plt.plot(x, y)
plt.show()
```

```python
n = np.sin(x) + np.cos(y) + np.random.uniform(0, 1, x.shape)
plt.imshow(n)
plt.show
```

```python
plt.imshow(np.fft.fftn(n).astype(np.uint8))
plt.show()
```

#Inverse FFT

```python
n = np.zeros((200, 200),dtype=complex)
n[60:80, 20:40] = np.exp(1j*np.random.uniform(0, 2*np.pi,(20, 20)))

plt.imshow(np.fft.ifftn(n).real, cmap='coolwarm')
plt.show()
```
